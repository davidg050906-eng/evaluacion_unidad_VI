## **2. Agregar producto (4 pts)**

Crear la función `agregar_producto(inventario)` que:

1. Solicite al usuario el código del producto.

2. Use try–except para validar:
    - Que el código sea un entero.
    - Que no esté repetido (si está repetido → lanzar raise ValueError).

3. Solicite nombre, categoría, precio y cantidad.

4. Inserte el nuevo producto en el inventario.

> Debe usar al menos un método de listas o diccionarios.

```python
def agregar_producto(inventario):
    try:
        entrada_codigo = input("Ingrese el código del producto (entero): ")
        codigo = int(entrada_codigo)
        for producto in inventario:
            if producto["codigo"] == codigo:
                raise ValueError(f"El código {codigo} ya existe.")
        nombre = input("Nombre: ")
        categoria = input("Categoría: ")
        precio = float(input("Precio: "))
        cantidad = int(input("Cantidad: "))
        nuevo_item = {
            "codigo": codigo,
            "nombre": nombre,
            "categoria": categoria,
            "precio": precio,
            "stock": cantidad
        }
        inventario.append(nuevo_item)
        print("Producto agregado exitosamente.")
    except ValueError as e:
        print(f"Error: {e}")
```

## **3. Buscar producto (4 pts)**

Crear la función `buscar_producto(inventario)` que:

1. Solicite un código.
2. Maneje los posibles errores:

    - Código no entero → capturar con except.
    - Producto no encontrado → lanzar raise KeyError.

3. Si lo encuentra, debe devolver los datos del producto.
4. Debe demostrar el uso de **indexación**, **in** o **keys()**, según la estructura elegida.

```python
def buscar_producto(inventario):
    try:
        entrada = input("Ingrese el código a buscar: ")
        codigo_buscado = int(entrada)
        producto_encontrado = None
        for producto in inventario:
            if "codigo" in producto:
                if producto["codigo"] == codigo_buscado:
                    producto_encontrado = producto
                    break
        if producto_encontrado is None:
            raise KeyError(f"El producto con código {codigo_buscado} no existe.")
        return producto_encontrado
    except ValueError:
        print("Error: El código debe ser un número entero.")
```

## **4. Estadísticas del inventario (4 pts)**

1. Cantidad total de productos distintos.
2. Número total de unidades en stock.

4. Una lista de comprensión con los productos cuyo stock sea mayor a 5.
5. Un conjunto (set) con todas las categorías sin repetir.

> Debe manejar excepciones internas si el inventario está vacío.

```python
def estadisticas(inventario):
    try:
        if len(inventario) == 0:
            raise Exception("Inventario vacío.")
        total_distintos = len(inventario)
        unidades_totales = sum(p['stock'] for p in inventario)
        promedio = sum(p['precio'] for p in inventario) / total_distintos
        stock_mayor_5 = [p['nombre'] for p in inventario if p['stock'] > 5]
        categorias_unicas = {p['categoria'] for p in inventario}
        print(f"1. Productos distintos: {total_distintos}")
        print(f"2. Unidades totales: {unidades_totales}")
        print(f"3. Precio promedio: ${promedio:.2f}")
        print(f"4. Stock > 5: {stock_mayor_5}")
        print(f"5. Categorías: {categorias_unicas}")
        return {"total": total_distintos, "stock": unidades_totales, "promedio": promedio, "filtro": stock_mayor_5, "categorias": categorias_unicas}
    except Exception as e:
        print(f"Error: {e}")
```
